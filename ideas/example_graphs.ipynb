## Idee 1

Hier beginnen we dan met de vertragingen over de jaren heen te tonen waar elke lijn een andere soort trein voorstelt.
- linechart
- interactief: zelf kiezen welke soorten treinen er getoond worden en niet

Een woordje uitleg over wat elk soort type trein voorstelt hier als inleiding voor de grafiek gebruikt worden.

```python
import random

# maken van fake data als voorbeeld
train_types = ["IC", "ICT", "INT", "L", "P", "S"]
months = ["Jan", "Mei", "Sept"]

def make_random_time_delay():
    timeline = []
    delays = []

    for i in range(2016, 2018):
        for month in months:
            timeline.append(month + " " + str(i))
            delays.append(random.randint(5000, 6000))
    
    return timeline, delays
```

```python
import matplotlib.pyplot as plt

# maken van de voorbeeld line chart, hier met alle treinen standaard op (zonder interactiviteit)
for train_type in train_types:
    timeline, delays = make_random_time_delay()
    plt.plot(timeline, delays, label=train_type)

plt.title('Soorten treinen en hun vertragingen')
plt.ylabel("vertraging in minuten")
plt.legend(loc="best")
plt.show()
```

## Idee 3

```python
import matplotlib.pyplot as plt
```

```python
# fake data
months = ["Jan", "Feb", "Ma", "Apr", "Mei", "Jun", "Jul", "Aug", "Sep", "Okt", "Nov", "Dec"]
values_total = [1000, 1233, 2400, 980, 1100, 2500, 5500, 7400, 2000, 1145, 842, 1344]
```

```python
cumulative = [values_total[0]]
for i in range(1, len(values_total)):
    cumulative.append(cumulative[i - 1] + values_total[i])


plt.fill_between(months, cumulative)
plt.title("Cumulatieve evolutie vertraging voor IC treinen over 1 jaar")
plt.ylabel("Vertraging in minuten")
plt.xlabel("Maand")
```

## Idee 4

Hier willen we de gemiddelde vertraging per maand voorstellen bij de treinen met een dot plot, misschien is er een algemene tendens dat de zomermaanden slechter zijn dan winter bvb. We kunnen verschillende zaken bekijken:
- over elk jaar heen de som nemen om zo de vertraging te bekijken van de gemiddelde maand van 1 trein
- alle treinen samen de gemiddelde vertraging van een bepaalde periode

```python
# fake data
def make_mean_delay():
    mean_delay = []
    for _ in range(3):
        mean_delay.append(random.randint(5000, 8000))
    
    return mean_delay
```

```python
plt.scatter(make_mean_delay(), months, label=train_type, color='b')

plt.title('Gemiddelde vertraging per maand')
plt.xlabel("vertraging in minuten")
plt.show()
```
